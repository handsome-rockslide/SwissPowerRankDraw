# Swiss draw tournament

This is a script that runs a swiss draw tournament, each time you hit run it will 
1. read in the tournament spreadsheet
2. display the team rankings
3. create the next round. 
4. It will then write the next round to the spreadsheet so that you can easily add the next set of results.

At the moment it doesn't provide feilds
Any changes that you wish to make to the tournament can be changed in the spreadsheet

The spreadsheet must have the following columns
team_name_x, team_x_score, team_name_y, team_y_score

```python
import pandas as pd
import numpy as np
from mwmatching import maxWeightMatching
from datetime import datetime
from sprd import *
```

```python

spread_sheet_name = 'Example.xlsx'
sheet_name = 'Sheet1'

#Do you want to overwrite? If so be away that any other sheets that you have may disappear
overwrite = False

original_df = pd.read_excel(spread_sheet_name, sheet_name = sheet_name)

results = RunRound(original_df, True)

if overwrite:
    results.to_excel(spread_sheet_name, sheet_name=sheet_name, index=False)
else:
    new_name = spread_sheet_name[:-5] + datetime.now().strftime("%Y%m%d_%H%M%S") + spread_sheet_name[-5:]
    results.to_excel(new_name, sheet_name=sheet_name, index=False)
    
```
